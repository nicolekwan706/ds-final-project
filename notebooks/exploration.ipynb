```python
#load the dataset

import pandas as pd

df = pd.read_csv("etl_cleaned_dataset.csv")

#basic shape and first few rows
print("Rows, columns:", df.shape)
df.head()
```

```python
#inspect columns, types, and missingness

df.columns
df.info()
df.isna().mean().sort_values(ascending=False)
```

```python
#rating distribution
for col in ["rating", "rated"]:
    if col in df.columns:
        print(f"\nValue counts for {col}:")
        print(df[col].value_counts().head(20))
```

```python
#top genres

for col in ["genres", "genre"]:
    if col in df.columns:
        print(f"\nSample genres from column: {col}")
        print(df[col].value_counts().head(20))
```

```python
#language and country distributions (top 10)
for col in ["language", "original_language", "country", "production_countries"]:
    if col in df.columns:
        print(f"\nTop values for {col}:")
        print(df[col].value_counts().head(10))
```
